```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/shark-tank-india-season-1-complete/Shark_Tank_India_S1.csv')
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
#Present
ama_present = len(df[df.aman_present == 1])
anu_present = len(df[df.anupam_present == 1])
ash_present = len(df[df.ashneer_present == 1])
gha_present = len(df[df.ghazal_present == 1])
nam_present = len(df[df.namita_present == 1])
pey_present = len(df[df.peyush_present == 1])
vin_present = len(df[df.vineeta_present == 1])
```

```python
plt.pie([ama_present, anu_present, ash_present, gha_present, nam_present, pey_present, vin_present], 
        labels = ['Aman','Anupam', 'Ashneer', 'Ghazal', 'Namita', 'Peyush', 'Vineeta'],
        autopct = '%0.1f%%',
        radius = 1.5)
plt.suptitle('Percentage of Judges present in the show', x = 0.55, y = 1.15, color = 'red')
plt.show()
```

```python
#Invested
ama_invested = len(df[df.aman_invested == 1])
anu_invested = len(df[df.anupam_invested == 1])
ash_invested = len(df[df.ashneer_invested == 1])
gha_invested = len(df[df.ghazal_invested == 1])
nam_invested = len(df[df.namita_invested == 1])
pey_invested = len(df[df.peyush_invested == 1])
vin_invested = len(df[df.vineeta_invested == 1])
```

```python
plt.pie([ama_invested, anu_invested, ash_invested, gha_invested, nam_invested, pey_invested, vin_invested], 
        labels = ['Aman','Anupam', 'Ashneer', 'Ghazal', 'Namita', 'Peyush', 'Vineeta'],
        autopct = '%0.1f%%',
        explode = [0.25,0,0,0,0,0,0],
        radius = 1.5)
plt.suptitle('Percentage of Judges invested', x = 0.55, y = 1.15, color = 'red')
plt.show()
```

```python
#Percent of investments
ama_percent = (ama_invested/ama_present)*100
anu_percent = (anu_invested/anu_present)*100
ash_percent = (ash_invested/ash_present)*100
gha_percent = (gha_invested/gha_present)*100
nam_percent = (nam_invested/nam_present)*100
pey_percent = (pey_invested/pey_present)*100
vin_percent = (vin_invested/vin_present)*100
```

```python
x_pos = np.arange(7)
sharks = ['Aman','Anupam', 'Ashneer', 'Ghazal', 'Namita', 'Peyush', 'Vineeta']
plt.xticks(x_pos, sharks)
plt.bar(x_pos ,[ama_percent, anu_percent, ash_percent, gha_percent, nam_percent, pey_percent, vin_percent])
plt.suptitle('Percentage of investments', color = 'red')
plt.tight_layout()
plt.show()
```
